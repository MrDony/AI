```python
class Rocket:
    def __init__(self, x, y, z, vx, vy, vz):
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz

    def get_position(self):
        return self.x, self.y, self.z

    def get_velocity(self):
        return self.vx, self.vy, self.vz

    def set_position(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def set_velocity(self, vx, vy, vz):
        self.vx = vx
        self.vy = vy
        self.vz = vz

rocket = Rocket(0, 0, 0, 10, 20, 30)
print("Position:", rocket.get_position())
print("Velocity:", rocket.get_velocity())

rocket.set_position(1, 2, 3)
rocket.set_velocity(11, 22, 33)

print("New Position:", rocket.get_position())
print("New Velocity:", rocket.get_velocity())
```

```python
%%time
import numpy as np
import matplotlib.pyplot as plt

# Define the two matrices of size 6x6
matrix1 = np.random.rand(6, 6)
matrix2 = np.random.rand(6, 6)
matrix1*=matrix2
# Extract two arrays from the matrices
array1 = matrix1[:, 1]
plt.plot(array1)
plt.xlabel('Index')
plt.ylabel('Result')
plt.title('Multiplication of two arrays')

array2 = matrix2[:, 3]
plt.plot(array2)
plt.xlabel('Index')
plt.ylabel('Result')
plt.title('Multiplication of two arrays')


# Multiply the two arrays element-wise
result = array1 * array2
# Plot the result using matplotlib
plt.plot(result, linestyle="dotted")
plt.xlabel('Index')
plt.ylabel('Result')
plt.title('Multiplication of two arrays')
plt.show()
```
